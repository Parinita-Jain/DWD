```python
#Select input and output from given dataset 
X=df_new.drop("price",axis=1) #select input
Y=df_new['price'] #select output
```

```python
Y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Apply Features scaling means apply StandardScaler  
#call inbuilt class StandardScaler 
from sklearn.preprocessing import StandardScaler  
```

```python
#create object of StandardScaler class
ss=StandardScaler()
#Apply StandardScaler on input variables X
#use fit_transform() inbuilt method of StandardScaler class
X_train=ss.fit_transform(X_train)  #apply scaling on training input
X_test=ss.transform(X_test) #apply scaling on testing input
```

```python
X_train
```

```python
#create a model with LinearRegression method
from sklearn.linear_model import LinearRegression
```

```python
#create the object of LinearRegression class
lr=LinearRegression()
```

```python
import os
os.getcwd()
```

```python
#train the model with 70% data
lr.fit(X_train,Y_train)
```

```python
#Test the model  with 30% data
Y_pred=lr.predict(X_test) #Y_pred=m1*X1+m2*X2+................+c
```

```python
#check slope /coefficient of each input 
lr.coef_
```

```python
X.columns
```

```python
dict={'Input Columns':X.columns,'Slope':lr.coef_}
df1=pd.DataFrame(dict)
df1
```

```python
#Check intercept
lr.intercept_
```

```python
#check error MSE
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print(mean_squared_error(Y_test,Y_pred))
```

```python
#find r2_score
print(r2_score(Y_test,Y_pred))
```

```python
#score got 78% which is good but not better
#Apply ANOVA (Analysis of Variance) Test : it is used to select important features to find the
#output (output is a contineous numeric)  . ANOVA test : is a feature selection method
#SelectKBest inbuilt class 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression  #if ANOVA TEst (F-test/F-score)
```

```python
#formula of F-Test/F-score =Variance between Group/VAriance within group 
#ANOVA test is also known as F-Test or F-statistics or F-score or F-regression 
```

```python
#Create object of SelectKBest class
#k means select inportant no. of inputs 
#here no. of inputs =14
#Create object of SelectKBest class
anova=SelectKBest(score_func=f_regression,k=6) #here k means no. of imp features k=6
X_train_imp=anova.fit_transform(X_train,Y_train)
#same testing input
X_test_imp=anova.transform(X_test)
#print(anova.get_support()) #return answer True/False 
#get_support() inbuilt method of SelectKBest class 
dict={'Input Columns':X.columns,'Support_Anova':anova.get_support()}
df2=pd.DataFrame(dict)
print(df2)
#Train the model with LinearRegression class 
#create object of LinearRegression class
lr=LinearRegression()
lr.fit(X_train_imp,Y_train)
print("Slope : ",lr.coef_)
print("Intercept : ",lr.intercept_)
#Check Testing Score
score=lr.score( X_test_imp,Y_test) #similar as r2_score
print("\nScore : ",round(score,2))
```
