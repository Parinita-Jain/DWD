```python
s="123this is 123 my number 09876"
pattern=re.compile(r"\d+")
match=re.match(pattern,s)
if match:
    print(f"The pattern found {match.group()}")
else:
    print("Pattern not found")
```

```python
pattern=re.compile("[a-zA-Z]+")
s="My phone number ; is : 12345"
result=re.findall(pattern,s)
result
```

```python
pattern=re.compile("[^1-3]+") # 
s="1234567890 hi there 12346789540"
result=re.findall(pattern,s)
result
```

```python
pattern=re.compile("[a-z]{2,6}")
s="My phone number ; is : 12345"
result=re.findall(pattern,s)
result
```

```python
pattern=re.compile("^a...s$")
s="aaafs"
match=re.match(pattern,s)
if match:
    print(f"Pattern found {match.group()}")
else:
    print("Pattern not found")
```

```python
pattern=re.compile("[6-9]\d{9}")
s="9789078901"
if len(s)==10:
    match=re.match(pattern,s)
    if match:
        print(f"Phone number found {match.group()}")
    else:
        print("Phone number not found")
else:
    print("Phone number not found")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
from bs4 import BeautifulSoup

```

```python
driver=webdriver.Chrome()
driver.get("https://www.amazon.in/s?bbn=1350380031&rh=n%3A1350380031%2Cp_n_age_range%3A1480708031&dc&pf_rd_i=1350380031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=7e4e962c-058a-4cc8-842d-131a59989416&pf_rd_r=H2E5EE6EMC0QKPH6HN4T&pf_rd_s=merchandised-search-4&qid=1675423879&rnid=1480704031&ref=lp_1350381031_nr_p_n_age_range_3")
```

```python
driver.page_source
```

```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
product_span_tag=soup.find_all("span",attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
```

```python
product_name=[p.text for p in product_span_tag]
product_name
```

```python
price_span_tag=soup.find_all("span",attrs={"class":"a-price-whole"})
```

```python
product_price=[p.text.replace(",","") for p in price_span_tag]
product_price
```

```python
rating_span_tag=soup.find_all("span",attrs={"class":"a-icon-alt"})
product_rating=[r.text[:3] for r in rating_span_tag]
product_rating
```

```python
print(len(product_name),len(product_price),len(product_rating))
```

```python
product_rating=product_rating[:30]
len(product_rating)
```

```python
import pandas as pd
df=pd.DataFrame({
    "Product_name":product_name,
    "Product_price":product_price,
    "Product_rating":product_rating
})
df.to_csv("C:\\Users\\dell\\Documents\\ParinitaJain\\ParinitaJain9_11\\amazon_toys_details.csv",
         index=False)
```
