```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
table_class = "sortable wikitable sticky-header static-row-numbers col1left col4left jquery-tablesorter"

response=requests.get(url)
print(response)
```

```python
soup=BeautifulSoup(response.text,"html.parser")
india_table=soup.find("table",{"class":"wikitable"})
india_table.prettify()
```

```python
df=pd.read_html(india_table.prettify())
df=df[0]
df
```
